```
import pandas as pd
```

```
df = pd.DataFrame({"id": ["ID1", "ID2", "ID3", "ID4", "ID5", "ID6", "ID7", "ID8", "ID9"],
                   "metro": ["MSA 1", "MSA 1", "MSA 1", "MSA 1", "MSA 1", "MSA 2", "MSA 2", "MSA 2", "MSA 2"],
                   "educ": ["BA", "BA", "BA", "HS", "HS", "BA", "BA", "HS", "HS"],
                   "sex": ["M", "F", "F", "F", "M", "M", "F", "F", "M"],
                   "question1": ["2 - Not likely", "2 - Not likely", "1 - Somewhat to very likely", "2 - Not likely", "1 - Somewhat to very likely", "2 - Not likely", "1 - Somewhat to very likely", "2 - Not likely", "2 - Not likely"],
                   "question2": ["2 - Not enough to eat", "1 - Enough to eat", "2 - Not enough to eat", "1 - Enough to eat", "2 - Not enough to eat", "1 - Enough to eat", "2 - Not enough to eat", "1 - Enough to eat", "2 - Not enough to eat"],
                   "weight": [0.8, 1.3, 2.5, 1.4, 0.7, 0.9, 1.2, 1.1, 0.9]})
```

```
# generate the topline column
df['topline'] = 'Topline'
df
```

```
df1 = df.melt(id_vars=['id','metro','topline','educ','sex','weight'], 
              value_vars=['question1', 'question2'], 
              var_name='question_group', 
              value_name='question_val')
df1
```

```
df2 = df1.melt(id_vars=['id','metro','question_group','question_val','weight'], 
              value_vars=['topline','educ','sex'], 
              var_name='xtab_group', 
              value_name='xtab_val') 	

df2
```

```
df3 = df2.groupby(['metro', 'question_group', 'question_val', 'xtab_group', 'xtab_val']).agg({'weight': 'sum'}).reset_index()

df3["weight_total"] = df3.groupby(['metro','question_group','xtab_group'])['weight'].transform('sum')
df3["share"] = df3.weight/df3.weight_total

df3["weight_total_val"] = df3.groupby(['metro','question_group','xtab_group','xtab_val'])['weight'].transform('sum')
df3["share_val"] = df3.weight/df3.weight_total_val
df3.sort_values(by= ['metro', 'question_group', 'xtab_group'], inplace=True, ascending=True)
df3
```
